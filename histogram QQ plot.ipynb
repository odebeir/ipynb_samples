```python
def histos(ima):
    h,bin = np.histogram(ima[:],range(257))
    cumh = np.cumsum(h)
    norm_cumh = 1.*cumh/cumh[-1]
    f = interp1d(norm_cumh, bin[:-1],bounds_error=False)       
    return h,norm_cumh,f
```

# Image QQ plot

```python
from numpy import sort
from numpy.random import permutation

def QQ_correct(ref,g,N=1000):
    N = 1000
    sample_g = sort(permutation(ref.flatten())[:N])
    sample_g2 = sort(permutation(g.flatten())[:N])

    z = np.polyfit(sample_g2, sample_g, 2)
    p = np.poly1d(z)
    xp = np.linspace(0, 255,100)
    plt.figure()
    plt.plot(sample_g2,sample_g,lw=2)
    plt.plot(xp, p(xp))
    # image correction
    r = p(g)
    h = histos(r)
    plot_histos(h,r)
```

```python
QQ_correct(g,lena)
```
